```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.models import Sequential
from keras.layers.recurrent import LSTM
from keras.layers.core import Dense, Activation, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from sklearn.utils import shuffle
from tensorflow import keras
%matplotlib inline

# For better visualization of the plots
from jupyterthemes import jtplot
jtplot.style('onedork')
```

```python
df = pd.read_csv("Data/monthly-milk-production-pounds-p.csv")
df = df.drop(df.index[len(df)-1])
df.columns = ['Month', 'Monthly milk production']
df.head()
```

```python
df.index = pd.to_datetime(df['Month'], format="%Y-%m")
df.drop(['Month'], axis=1, inplace=True)
df.plot(x=df.index, y='Monthly milk production')
```

```python
train = df[df.index < '1975-01-01']
test = df[df.index >= '1975-01-01']
```

```python
scaler = MinMaxScaler()
train.loc[:,'Monthly milk production'] = scaler.fit_transform(train)
test.loc[:,'Monthly milk production'] = scaler.transform(test)
train.head()
```

## Building Computational Graph

```python
# Training Parameters
learning_rate = 0.001
training_steps = 5000
batch_size = 12
display_step = 200

# Network Parameters
num_input = 1
timesteps = 10 # timesteps
num_hidden = 64 # hidden layer num of features
num_classes = 1 

# tf Graph input
X = tf.placeholder("float", [None, timesteps, num_input])
Y = tf.placeholder("float", [None, timesteps, num_classes])
```

```python
def get_next_batch(training_data, seed, batch_size, pred_delta): # pred_steps needs to be less than batch_size
    """
    INPUT: Data, Batch Size, Time Steps per batch
    OUTPUT: A tuple of y time series results. y[:,:-1] and y[:,1:]
    """
    # STEP 1: Set the seed
    steps = batch_size - pred_delta
    start = seed
    # STEP 2: Now that you have a starting index you'll need to index the data from
    # the random start to random start + steps + 1. Then reshape this data to be (1,steps+1)
    y = training_data.iloc[start:start+batch_size].values.reshape(-1, batch_size)
    # STEP 3: Return the batches. You'll have two batches to return y[:,:-1] and y[:,1:]
    # You'll need to reshape these into tensors for the RNN to .reshape(-1,steps,1)
    return y[:, :-pred_delta].reshape(-1,steps,1), y[:, pred_delta:].reshape(-1,steps,1)

seed = 0
pred_delta = 2
for i in range(100):
    x, y = get_next_batch(train, seed, 12, pred_delta)
    print (i, x.shape, y.shape)
    seed += 12
    
    if seed + batch_size > len(train):
        print ("1 iteration complete")
        break
```

```python
def RNN(x, num_hidden, weights, biases):
    #x = tf.unstack()
    lstm_cell = tf.contrib.rnn.OutputProjectionWrapper(tf.contrib.rnn.BasicLSTMCell(num_units=num_hidden,activation=tf.nn.relu),
                                                  output_size=num_classes, reuse=tf.AUTO_REUSE)
    outputs, states = tf.nn.dynamic_rnn(lstm_cell, x, dtype=tf.float32)
    return outputs
#     lstm_cell = tf.contrib.rnn.BasicLSTMCell(num_hidden)
#     outputs, states = tf.contrib.rnn.static_rnn(lstm_cell, x, dtype=tf.float32)
#     return tf.matmul(outputs, weights['out']) + biases['out']
```

```python
predictions = RNN(X, num_hidden, weights=0, biases=0)
loss = tf.reduce_mean(tf.square(predictions - Y)) # Mean-Squared error
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)
train_op = optimizer.minimize(loss)
init = tf.global_variables_initializer()
```

```python
with tf.Session() as sess:
    sess.run(init)
    seed = 0
    for i in range(training_steps):
        #print (seed)
        batch_x, batch_y = get_next_batch(train, seed, 12, pred_delta)
        sess.run(train_op, feed_dict={X:batch_x, Y:batch_y})
        seed += 12
        
        if i % 100 == 0 or i == 1:
            l = sess.run(loss, feed_dict={X:batch_x, Y:batch_y})
            print ("Iteration:", i, "\tMSE:", l)
        
        if seed + batch_size > len(train):
            seed = 0
```

## LSTM using Keras API

```python
def create_dataset(dataset, window_size = 1):
    data_X, data_Y = [], []
    for i in range(len(dataset) - window_size - 1):
        a = dataset[i:(i + window_size), 0]
        data_X.append(a)
        data_Y.append(dataset[i + window_size, 0])
    return(np.array(data_X), np.array(data_Y))
```

```python
# Create test and training sets for one-step-ahead regression.
window_size = 1
train_X, train_Y = create_dataset(train.values, window_size)
test_X, test_Y = create_dataset(test.values, window_size)
print("Original training data shape:")
print(train_X.shape)

# For the RNN, you'll need to reshape these into tensors to 
# .reshape(-1,steps,1) OR
# .reshape(data.shape[0],steps,data.shape[1])
train_X = np.reshape(train_X, (train_X.shape[0], 1, train_X.shape[1]))
test_X = np.reshape(test_X, (test_X.shape[0], 1, test_X.shape[1]))
print("New training data shape:")
print(train_X.shape)
```

```python
def init_model(train_X, train_Y, window_size = 1):
    
    model = keras.Sequential()
    model.add(keras.layers.LSTM(12, return_sequences=True, input_shape=(1, window_size)))
    model.add(keras.layers.Dropout(0.5))
    model.add(keras.layers.Conv1D(filters=32, kernel_size=2, padding='same', activation='relu'))
    #model.add(keras.layers.MaxPooling1D(pool_size=2))
    model.add(keras.layers.LSTM(12, return_sequences=False))
    model.add(keras.layers.Dropout(0.5))
    model.add(keras.layers.Flatten())
    model.add(keras.layers.Dense(1))
    model.add(keras.layers.Activation('linear'))

    model.compile(optimizer='adam',
                  loss='mean_squared_error',
                  metrics = ['mse'])
    print (model.summary())
    return model

model = init_model(train_X, train_Y, window_size = 1)
```

```python
model.fit(train_X,
              train_Y,
              epochs = 100,
              batch_size = 2)
```
