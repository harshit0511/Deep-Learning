```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import cricpy.analytics as ca
```

```python
kohli =ca.getPlayerData(253802,dir="./",file="kohli.csv",type="batting",homeOrAway=[1,2], result=[1,2,4])
# Converting to datetime and making it the index
kohli.index = pd.to_datetime(kohli['Start Date'],infer_datetime_format=True)
# Dropping old Dates column
kohli.drop(['Start Date'], axis=1, inplace=True)
# Removing * from the end of runs (eg. 127*)
kohli['Runs'] = kohli['Runs'].str.split('*').str[0]
# Converting 'DNB' etc. to NaN
kohli = kohli.convert_objects(convert_numeric=True)
# Replacing NaN values with 0
kohli.fillna(0, inplace=False)
kohli.head()
```
